### Plot all valid USGS sites

```python
df = pd.read_excel(f'Prelim_Data\\National_Validity.xlsx')

# Filter sites
df_not_val = df.loc[(df['data_range'] == False) & (df['data_cont'] == False)]
df_val_range = df.loc[(df['data_range'] == True)  & (df['data_cont'] == False)]
df_val = df.loc[(df['data_range'] == True) & (df['data_cont'] == True)]
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,10))

# Plot us49
us49.plot(ax=ax, edgecolor='black', alpha = 0.2, legend=True) 

# Plot points
def plot_points(df):
    lat = df['dec_lat_va'].tolist()
    long = df['dec_long_va'].tolist()
    geometry = [Point(xy) for xy in zip(long, lat)]
    geo_df = gpd.GeoDataFrame(geometry=geometry)
    return geo_df

geo_not_val = plot_points(df_not_val)
geo_val_range = plot_points(df_val_range)
geo_val = plot_points(df_val)

bx = geo_not_val.plot(ax=ax, color='red', marker='o', markersize=10, edgecolor='red', linewidth=1, alpha=0.75, legend=True)
bx = geo_val_range.plot(ax=ax, color='yellow', marker='o', markersize=10, edgecolor='yellow', linewidth=1, alpha=0.75, legend=True)
bx = geo_val.plot(ax=ax, color='green', marker='o', markersize=10, edgecolor='green', linewidth=1, alpha=0.75, legend=True)

print(f'All points: {len(df)}')
print(f'Not valid points: {len(df_not_val)}')
print(f'Valid points (data range only): {len(df_val_range)}')
print(f'Valid points: {len(df_val)}')
print(f'Percent of valid sites: {len(df_val)/len(df)*100: 0.1f}%')
```
